```python
import unicodecsv

enrollments = []
with open('enrollments.csv', 'rb') as f :  #python csv library document says you should read files in 'rb'  
    reader = unicodecsv.DictReader(f)

    print('--------reader : DictReader Class is iterable --------')
    print(reader)

    for row in reader:
        enrollments.append(row)
    
print('--------enrollments----------')
enrollments[0] 
```

# 04 Fixing Data Types 
<hr/>

Kind of Data Cleaning 

```python
## continue with ... enrollments.csv

from datetime import datetime as dt
def parse_date(date) : 
    if date == '' : 
        return None 
    else : 
        return dt.strptime(date , '%Y-%m-%d')
    
def parse_int(i) : 
    if i == '' : 
        return None 
    else : 
        return int(i)

    
for e in enrollments : 
    e['cancel_date']  = parse_date(e['cancel_date'])
    e['join_date']  = parse_date(e['join_date'])
    e['days_to_cancel'] = parse_int(e['days_to_cancel'])
    e['is_canceled'] = (e['is_canceled'] == 'True')
    e['is_udacity'] = (e['is_udacity'] == 'True' )

enrollments[0]
```

```python
### continue with ... daily_engagement.csv 
'''
{'acct': '0',
 'lessons_completed': '0.0',
 'num_courses_visited': '1.0',
 'projects_completed': '0.0',
 'total_minutes_visited': '11.6793745',
 'utc_date': '2015-01-09'}
'''

from datetime import datetime as dt
def parse_date(date) : 
    if date == '' : 
        return None 
    else : 
        return dt.strptime(date , '%Y-%m-%d')
    
def parse_float(f) : 
    if f == '' : 
        return None 
    else : 
        return int(float(f))

    
for e in engagement_list : 
    e['lessons_completed'] =  parse_float(e['lessons_completed'])
    e['num_courses_visited'] = parse_float(e['num_courses_visited'])
    e['projects_completed'] = parse_float(e['projects_completed'])
    e['total_minutes_visited'] = float(e['total_minutes_visited'])
    e['utc_date'] = parse_date(e['utc_date'])
    
engagement_list[0]
```

```python
surprisings =  unique_enrolled_students - unique_engagement_students

surprizing = []
for enrollment in enrollments : 
    student = enrollment['account_key'] 
    if student not in unique_engagement_students : 
        surprizing.append(enrollment)

canceled= []
for enrollment in enrollments : 
    if enrollment['is_canceled'] : 
        canceled.append(enrollment)
        
for c in canceled : 
    if c not in surprizing : 
        print(c)
for i in range(9) : 
    print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
print(surprizing)

print(len(surprizing))
for s in surprizing : 
    if s['days_to_cancel'] != 0 : 
        print(s)
```

```python
udacity_testcase = set(e['account_key'] for e in enrollments if e['is_udacity'] )

def non_udacity_dataset(dataset) : 
    result = []
    for data in dataset : 
        if data['account_key'] not in udacity_testcase  : 
            result.append(data)
    return result

    
non_udacity_enrollments = non_udacity_dataset(enrollments)
non_udacity_engagements = non_udacity_dataset(engagement_list)
non_udacity_submissions = non_udacity_dataset(submissions_list)

non_udacity_enrollments[0]
```

```python
paid_students = dict()
for student in non_udacity_enrollments :
    if (student['days_to_cancel']==None) or (student['days_to_cancel'] > 7 ):
        #print(student)
        account_key = student['account_key']
        enrollment_date = student['join_date']
        if (account_key not in paid_students) or (enrollment_date > paid_students[account_key]) :
            paid_students[account_key] = enrollment_date 

len(paid_students)
print(paid_students['300'])
```

# 12. Getting Data from First week
<hr/> 

First, create list of data  only by students in paid_students . 

Then, find the engagement records containing only data for paid students during ** their first week ** 


```python
#First

def paid_dataset(dataset)  :
    result = list()
    for data in dataset : 
        if data['account_key'] in paid_students : 
            result.append(data)
    return result


paid_enrollments = paid_dataset(non_udacity_enrollments)
paid_engagements = paid_dataset(non_udacity_engagements)
paid_submissions = paid_dataset(non_udacity_submissions)

print(paid_engagements[0])
len(paid_engagements)
```

```python
#Then

def within_one_week(join_date, engagement_date):
    time_delta = engagement_date - join_date
    return (time_delta.days < 7 ) & (time_delta.days>=0)


paid_engagement_in_first_week = list()

for data in paid_engagements : 
    account_key = data['account_key'] 
    engagement_date = data['utc_date']  
    if within_one_week(paid_students[account_key], engagement_date) : 
        paid_engagement_in_first_week.append(data)
        
print(paid_engagement_in_first_week[0])
print(len(paid_engagement_in_first_week))
```

# 13. Exploring Data : Exploring Student Engagment
<hr/>
Arrange `paid_engagement_in_first_week` by each student <br/> 
and compare the average engaging time of each student. 

```python
from collections import defaultdict

def group_data(dataset, key_name) :
    result = defaultdict(list)
    for data in dataset : 
        key = data[key_name] 
        result[key].append(data)
    return result

engagement_by_account = group_data(paid_engagement_in_first_week, 'account_key') #키의 default value 값을 list 로 지정
    
engagement_by_account['100']
```

```python
def sum_grouped_items(grouped_data, field_name) :
    result = dict()
    for account_key, datas in grouped_data.items() : 
        total = 0 
        for data in datas :
            total += data[field_name]
        result[account_key] = total 
    return result

total_minute_by_account = sum_grouped_items(engagement_by_account, 'total_minutes_visited')

total_minute_by_account
```

```python
import numpy as np

def describe_data(sumed_grouped_items) : 
    totals = list(sumed_grouped_items.values())

    print("mean:",np.mean(totals))
    print("standard deviation:",np.std(totals))
    print("max:",np.max(totals))
    print("min:",np.min(totals))
    
describe_data(total_minute_by_account)
```

```python
def sum_on_off(grouped_data, field_name) :
    result = dict()
    for account_key, datas in grouped_data.items() : 
        total = 0 
        for data in datas :
            if data[field_name] > 0 : 
                total += 1 
        result[account_key] = total 
    return result 

days_visited_by_account = sum_on_off(engagement_by_account, 'num_courses_visited')
describe_data(days_visited_by_account)
```
